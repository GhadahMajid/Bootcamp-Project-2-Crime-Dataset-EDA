## Is there any missing values?

```python
fig, ax = plt.subplots(figsize=(9,4))
sns.heatmap(df.isnull(), cbar=False, cmap="YlGnBu_r")
plt.title('The Missing Values Chart')
plt.xlabel('Atrributes',size=15)
plt.ylabel('# Nulss',size=15);
plt.show()
```

## Drop irrelevant data

```python
df = df.drop(columns=['CR Number','Offence Code','NIBRS Code','Dispatch Date / Time','Zip Code', 'Sector','Beat','PRA','Location','Street Prefix','Street Suffix','Address Number','Latitude','Longitude','Block Address','Crime Name3','Crime Name1'], axis=1)
df.head()
```

## Filling the null values

```python
df['End_Date_Time']=df['End_Date_Time'].fillna('countinued')
```

```python
df['City']=df['City'].fillna(df['City'].mode()[0])
```

```python
df['Street Type']=df['Street Type'].fillna(df['Street Type'].mode()[0])
```

```python
fig, ax = plt.subplots(figsize=(9,4))
sns.heatmap(df.isnull(), cbar=False, cmap="YlGnBu_r")
plt.title('The Missing Values Chart')
plt.xlabel('Atrributes',size=15)
plt.ylabel('# Nulss',size=15);
plt.show()
```
