## Is there any missing values?

```python
fig, ax = plt.subplots(figsize=(9,4))
sns.heatmap(df.isnull(), cbar=False, cmap="YlGnBu_r")
plt.title('The Missing Values Chart')
plt.xlabel('Atrributes',size=15)
plt.ylabel('# Nulss',size=15);
plt.show()
```

```python
df.info(); #null: Crime Name2,Police District Name, Street Name
```

## Drop irrelevant data

```python
df.drop(columns=['CR Number','Offence Code','NIBRS Code','Dispatch Date / Time','Zip Code', 'Sector','Beat','PRA','Location','Street Prefix','Street Suffix','Address Number','Latitude','Longitude','Block Address'], axis=1, inplace=True)
df.head()
```

## Filling the null values

```python
df['End_Date_Time']=df['End_Date_Time'].fillna('countinued')
```

```python
df['City']=df['City'].fillna(df['City'].mode()[0])
```

```python
df['Street Type']=df['Street Type'].fillna(df['Street Type'].mode()[0])
```

```python
fig, ax = plt.subplots(figsize=(9,4))
sns.heatmap(df.isnull(), cbar=False, cmap="YlGnBu_r")
plt.title('The Missing Values Chart')
plt.xlabel('Atrributes',size=15)
plt.ylabel('# Nulss',size=15);
plt.show()
```

Data Visualization

```python
#1
plt.bar(df['Year'].unique(),df['Year'].value_counts())
plt.title('crime frequency by the years',size=15)
plt.xlabel('Year Of Crime',size=15)
plt.ylabel('Number Of Crimes',size=15);
```

```python
#2
plt.pie(df['City'].value_counts(), labels=df['City'].unique(), explode=np.linspace(0.4,0.4,70));
plt.title('crimes by the city\n',size=15);
```

```python
#3
#count of crime group by type
group = df[['Victims', 'Crime Name1']].groupby('Crime Name1').count()
sns.heatmap(group);
plt.title("count of victims grouped by crime type");
plt.ylabel('Crimes');
```

```python
#4
#Embezzlement crime on period (2020-2021-2022)
group =df[(df['Year'] =='2020')&(df['Crime Name2']=='Embezzlement')]
group = group[['Victims', 'Crime Name3']].groupby('Crime Name3').count()

group2 =df[(df['Year'] =='2021')&(df['Crime Name2']=='Embezzlement')]
group2 = group2[['Victims', 'Crime Name3']].groupby('Crime Name3').count()

group3 =df[(df['Year'] =='2022')&(df['Crime Name2']=='Embezzlement')]
group3 = group3[['Victims', 'Crime Name3']].groupby('Crime Name3').count()


plt.figure(figsize=(20,8))
#sns.countplot(x='Victims', data=group);



plt.plot(group['Victims'], linestyle = '--', linewidth = 3);
plt.plot(group2['Victims'], linestyle = '--', linewidth = 3);
plt.plot(group3['Victims'], linestyle = '--', linewidth = 3);

plt.legend(['2020','2021'], shadow = True, loc = 0);
plt.title("Embezzlement crime on period (2020-2021-2022)");
plt.xlabel('Embezzlement Crimes');
plt.ylabel('Sales');
```

```python
#5 
#Abduction crimes in (2020-2021-2022)
group =df[(df['Year'] >='2020')&(df['Crime Name2']=='Kidnapping/Abduction')]
#
plt.figure(figsize=(20,8))
sns.histplot(
    group,
    x="Crime Name3",
    hue="Year",
    multiple="stack", 
    alpha=0.8);

plt.title("Abduction crimes in (2020-2021-2022)");
plt.xlabel('Abduction Crimes');
plt.ylabel('Incidents');
```
