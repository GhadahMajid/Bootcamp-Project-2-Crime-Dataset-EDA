## Dealing with null values

```python
df.isnull().sum()
```

```python
df = df[df['Crime Name2'].notna()]
```

```python
df['Police District Name'].value_counts()
```

```python
df['Police District Name'] = df['Police District Name'].fillna('OTHER')
```

```python
df = df[df['Crime Name2'].notna()]
```

```python
df['City']=df['City'].fillna(df['City'].mode()[0])
```

```python
df['Street Type']=df['Street Type'].fillna(df['Street Type'].mode()[0])
```

```python
df.isnull().sum().sum()
```

## The datafarme after dealing the nulls

```python
df.head()
```

```python
df['Agency'].value_counts()
```

```python
df['Agency'].unique()
```

```python
df['State'].value_counts()
```

```python
df['Crime Name2'].value_counts()
```

```python
df['Year']=df['Start_Date_Time'].str.slice(6,10)
```

```python
df['Crime Time'] = pd.to_datetime(df['Start_Date_Time']).dt.hour
```

```python
df['Crime Period'] =  df.apply(lambda row: 
                                 'Morning' if (row['Crime Time'] < 12) 
                                  else ('Afternoon' if row['Crime Time'] >= 12 and row['Crime Time'] <=16
                                  else 'Night'), axis=1)
```

```python
df['Case File'] = df.apply(lambda row: 
                                 'Open' if (pd.isnull(row['End_Date_Time'])) 
                                  else 'Closed', axis=1)
```

```python
df['Case File'].value_counts()
```

```python
df.drop(['Start_Date_Time','Crime Time','End_Date_Time'],axis=1, inplace = True)
```

```python
#4
#count of crime group by type
group = df[['Victims', 'Crime Name1']].groupby('Crime Name1').count()
sns.heatmap(group);
plt.title("count of victims grouped by crime type");
plt.ylabel('Crimes');
```

```python
#5
#Embezzlement crime on period (2020-2021-2022)
group =df[(df['Year'] =='2020')&(df['Crime Name2']=='Embezzlement')]
group = group[['Victims', 'Crime Name3']].groupby('Crime Name3').count()

group2 =df[(df['Year'] =='2021')&(df['Crime Name2']=='Embezzlement')]
group2 = group2[['Victims', 'Crime Name3']].groupby('Crime Name3').count()

group3 =df[(df['Year'] =='2022')&(df['Crime Name2']=='Embezzlement')]
group3 = group3[['Victims', 'Crime Name3']].groupby('Crime Name3').count()


plt.figure(figsize=(20,8))
#sns.countplot(x='Victims', data=group);

plt.plot(group['Victims'], linestyle = '--', linewidth = 3);
plt.plot(group2['Victims'], linestyle = '--', linewidth = 3);
plt.plot(group3['Victims'], linestyle = '--', linewidth = 3);

plt.legend(['2020','2021','2022'], shadow = True, loc = 0);
plt.title("Embezzlement crime on period (2020-2021-2022)");
plt.xlabel('Embezzlement Crimes');
plt.ylabel('Number of crime');
```

```python
#6
#Abduction crimes in (2020-2021-2022)
group =df[(df['Year'] >='2020')&(df['Crime Name2']=='Kidnapping/Abduction')]
#
plt.figure(figsize=(20,8))
sns.histplot(
    group,
    x="Crime Name3",
    hue="Year",
    multiple="stack", 
    alpha=0.8)

plt.title("Abduction crimes in (2020-2021-2022)")
plt.xlabel('Abduction Crimes')
plt.ylabel('Incidents')
plt.xticks(rotation=90);
```
